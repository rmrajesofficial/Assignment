```python
X=df.drop(["phishing"],axis=1)
Y=df["phishing"]
X.shape
```

```python
Y.shape
```

```python
from sklearn.model_selection import train_test_split
from sklearn.neighbors import KNeighborsClassifier
from sklearn.svm import SVC
from sklearn.linear_model import LogisticRegression
from sklearn.preprocessing import StandardScaler

from sklearn.metrics import accuracy_score, r2_score, mean_squared_error
```

```python
X_train,X_test,y_train,y_test = train_test_split(X,Y,test_size=0.2, random_state=42)
```

```python
data={"name":[],
      "accuracy":[],
      "accuracy_scaled":[],
      "r2":[],
      "r2_scaled":[],
      "mse":[],
      "mse_scaled":[]}
scaler=StandardScaler()

X_train_scaled=scaler.fit_transform(X_train)
X_test_scaled=scaler.transform(X_test)
for i in range(8,13):
    model1_instance = KNeighborsClassifier(n_neighbors=i)
    model2_instance = KNeighborsClassifier(n_neighbors=i)
    model1_instance.fit(X_train, y_train)
    model2_instance.fit(X_train_scaled, y_train)
    y_pred = model1_instance.predict(X_test)
    y_pred_scaled = model2_instance.predict(X_test_scaled)
    data["name"].append(f"KNN-{i}")
    data["accuracy"].append(accuracy_score(y_test, y_pred))
    data["accuracy_scaled"].append(accuracy_score(y_test, y_pred_scaled))
    data["r2"].append(r2_score(y_test, y_pred))
    data["r2_scaled"].append(r2_score(y_test, y_pred_scaled))
    data["mse"].append(mean_squared_error(y_test, y_pred))
    data["mse_scaled"].append(mean_squared_error(y_test, y_pred_scaled))

svm1 = SVC()
svm2 = SVC()
svm1.fit(X_train,y_train)
svm2.fit(X_train_scaled, y_train)
y_svm_pred=svm1.predict(X_test)
y_svm_pred_scaled=svm2.predict(X_test_scaled)
data["name"].append(f"SVM")
data["accuracy"].append(accuracy_score(y_test, y_svm_pred))
data["accuracy_scaled"].append(accuracy_score(y_test, y_svm_pred_scaled))
data["r2"].append(r2_score(y_test, y_svm_pred))
data["r2_scaled"].append(r2_score(y_test, y_svm_pred_scaled))
data["mse"].append(mean_squared_error(y_test, y_svm_pred))
data["mse_scaled"].append(mean_squared_error(y_test, y_svm_pred_scaled))


softmax1 = LogisticRegression(multi_class='multinomial', solver='lbfgs')
softmax2 = LogisticRegression(multi_class='multinomial', solver='lbfgs')
softmax1.fit(X_train, y_train)
softmax2.fit(X_train_scaled, y_train)
y_soft_pred=softmax1.predict(X_test)
y_soft_pred_scaled=softmax2.predict(X_test_scaled)

data["name"].append(f"Softmax")
data["accuracy"].append(accuracy_score(y_test, y_soft_pred))
data["accuracy_scaled"].append(accuracy_score(y_test, y_soft_pred_scaled))
data["r2"].append(r2_score(y_test, y_soft_pred))
data["r2_scaled"].append(r2_score(y_test, y_soft_pred_scaled))
data["mse"].append(mean_squared_error(y_test, y_soft_pred))
data["mse_scaled"].append(mean_squared_error(y_test, y_soft_pred_scaled))
```
